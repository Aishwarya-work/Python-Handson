* A RegEx, or Regular Expression, is a sequence of characters that forms a search pattern.

* RegEx can be used to check if a string contains the specified search pattern.

* Python has a built-in package called re, which can be used to work with Regular Expressions.

```python
import re
```

# FindAll ():

```python
txt = "The rain in Spain"
x = re.findall("ai", txt)
print(x)
```

```python
txt = "The rain in Spain"
x = re.findall("Portugal", txt)
print(x)
```

# FindSearch ()

```python

txt = "The rain in Spain"
x = re.search(r"\s", txt)

print("The first white-space character is located in position:", x.start())
```

# Split()

```python
txt = "The rain in Spain"
x = re.split(r"\s", txt)
print(x)
```

# Split() 

```python
txt = "The rain in Spain"
x = re.sub(r"\s", "*", txt)
print(x)
```

# Extracting Phone Numbers Using Regex in Python

```python
import re
```

```python
text = """
Call me at 9876543210 or at +91-9876543210.
Also reachable at (91) 98765-43210 or (987) 654 3210 or 1111111111. (999)-654-4232
"""
```

```python
pattern = r'\d{10}'     # {} -- exactly the specified number of occurences
matches = re.findall(pattern,text)
print(matches)
```

```python
#US PHONE NUMBER
pattern1 = r'\(\d{3}\)'
matches1 = re.findall(pattern1,text)
print(matches1)
```

```python
pattern2 = r'\(\d{3}\)-\d{3}-\d{4}'
matches2 = re.findall(pattern2,text)
print(matches2)
```

# Extract Indian and US Number

```python
pattern3 = r'\d{10}|\(\d{3}\)-\d{3}-\d{4}'     #   | -- OR OPERATOR 
matches = re.findall(pattern3,text)
print(matches)
```

# Extract financial period

```python
text = '''
The gross cost of operating lease vehicles in FY2021 Q1 was $4.85 billion.
In previous quarter i.e. FY2020 Q4 it was $3 billion.
 FY2021 Q5 was $4.85 billion.
'''
```

```python
pattern = r'FY\d{4} Q[1-4]'      #  
matches = re.findall(pattern,text) 
print(matches)
```

# case insensitive match using flags

```python
text = '''
The gross cost of operating lease vehicles in FY2021 Q1 was $4.85 billion.
In previous quarter i.e. fy2020 Q4 it was $3 billion.
 FY2021 Q5 was $4.85 billion.
'''
```

```python
pattern = r'FY\d{4} Q[1-4]'      #  
matches = re.findall(pattern,text,flags = re.IGNORECASE) 
print(matches)
```

```python
pattern = r'FY(\d{4} Q[1-4])'      #  
matches = re.findall(pattern,text,flags = re.IGNORECASE) 
print(matches)
```

# Using Regex on Pandas Dataframe

```python
import pandas as pd

data = pd.DataFrame({
    "Phone Text": [
        "Akash number is 9999999999",
        "Rahul number is 9980787866",
        "Venu number is 9908707866"
    ]
})
```

```python
data
```

# Pandas extract

```python
data["Phone Number"] = data["Phone Text"].str.extract(r'(\d{10})')
data.head()
```

# Using apply function

```python
def extract_ph_num(text):
    pattern = r'\d{10}'
    search_obj = re.search(pattern,text)
    number = search_obj.group()
    return number
```

```python
data["Apply Phone Number"] = data["Phone Text"].apply(extract_ph_num)
data.head()
```
